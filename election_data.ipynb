```python


# Grouping by Leading Party
grouped = sorted_df.groupby("Leading Party").agg({
    "Constituency": "count",  # Number of constituencies won
    "Margin": "sum"           # Total margin
}).rename(columns={"Constituency": "Constituencies Won", "Margin": "Total Margin"})

print(grouped)

# Save to CSV
grouped.to_csv("grouped_by_party.csv")
```

```python
grouped["Constituencies Won"].plot(kind="bar", figsize=(10, 6), color="skyblue")
plt.title("Constituencies Won by Party")
plt.xlabel("Party")
plt.ylabel("Number of Constituencies Won")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Const. No.'], y=df['Margin'])
plt.title('Margin of Victory vs Constituency Number')
plt.xlabel('Constituency Number')
plt.ylabel('Margin of Victory')
plt.grid(True)
plt.show()
```

```python
# Count the occurrences of leading and trailing candidates
leading_candidate_counts = df['Leading Candidate'].value_counts().head(10)
trailing_candidate_counts = df['Trailing Candidate'].value_counts().head(10)

# Plot top 10 leading candidates
plt.figure(figsize=(12, 6))
sns.barplot(x=leading_candidate_counts.index, y=leading_candidate_counts.values)
plt.title('Top 10 Leading Candidates')
plt.xlabel('Candidate')
plt.ylabel('Count of Constituencies')
plt.xticks(rotation=90)
plt.show()

# Plot top 10 trailing candidates
plt.figure(figsize=(12, 6))
sns.barplot(x=trailing_candidate_counts.index, y=trailing_candidate_counts.values)
plt.title('Top 10 Trailing Candidates')
plt.xlabel('Candidate')
plt.ylabel('Count of Constituencies')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plotting margin distribution for each leading party
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Margin', hue='Leading Party', kde=True, bins=20)
plt.title('Margin of Victory Distribution by Leading Party')
plt.xlabel('Margin of Victory')
plt.ylabel('Frequency')
plt.show()
```

```python
# Compare margin for leading vs trailing parties
plt.figure(figsize=(16, 8))
sns.boxplot(x='Leading Party', y='Margin', data=df)
plt.title('Margin of Victory by Leading Party')
plt.xticks(rotation=45, fontsize=10, ha='right') 
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Trailing Party', y='Margin', data=df)
plt.title('Margin of Victory by Trailing Party')
plt.xticks(rotation=90)
plt.show()
```
